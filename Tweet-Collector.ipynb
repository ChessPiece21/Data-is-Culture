# Get Tweets and & Metadata

To collect tweets and associated metadata, we can use the command [`twarc2 search`](https://twarc-project.readthedocs.io/en/latest/twarc2/#search) and insert a query. If you have an Academic Research account, you can collect tweets from the entire archive with the flag `--archive`.

```python
!twarc2 search "(\"puget sound basketball league\")" --archive --limit 30 --tweet-fields "text"
```

```python
import plotly.express as px
import pandas as pd

filename = "psbl_tweet_counts.csv"

plot_title = "Tweets"

# reading in CSV file
counts_df = pd.read_csv(filename, parse_dates=['start'])
# Make sure to sort from start to end
counts_df = counts_df.sort_values('start')

# Just plug in the dataframe and pick the x, y columns -- also change the labels for x, y axes
fig = px.line(
    counts_df, 
    x='start', 
    y='day_count',
    labels={'start': 'Time', 'day_count': 'Tweets per Day'},
    # You will notice that I'm obsessed with f-strings. Sorry!
    title= f'{plot_title}'
)
fig.show()
```

```python
counts_df.set_index('start', inplace=True)
```

```python
month_count = counts_df.resample('M')['day_count'].sum().reset_index()

fig = px.line(
    month_count, 
    x='start', 
    y='day_count',
    labels={'start': 'Time', 'day_count': 'Tweets per Day'},
    # You will notice that I'm obsessed with f-strings. Sorry!
    title= f'{plot_title}'
)
fig.show()
```

# Read in Tweet Data

```python
import pandas as pd
pd.options.display.max_colwidth = 600
pd.options.display.max_rows = 600

filename = "psbl_tweets.csv"
tweets_df = pd.read_csv(filename, parse_dates=['created_at'])
tweets_df
```

Look how many columns there are!

```python
tweets_df.info()
```

```python
tweets_df.rename(columns={'created_at': 'date',
                          'public_metrics.retweet_count': 'retweets', 
                          'author.username': 'username', 
                          'author.name': 'name',
                          'author.verified': 'verified', 
                          'public_metrics.like_count': 'likes', 
                          'public_metrics.quote_count': 'quotes', 
                          'public_metrics.reply_count': 'replies',
                           'author.description': 'user_bio'},
                 inplace=True)

tweets_df = tweets_df[['date', 'text', 'username', 'name',  'retweets',
           'in_reply_to_user_id', 'likes', 'replies',  'quotes',  'user_bio', 'verified', 'context_annotations']]
```
